```python
import pyvista as pv
import os
```

```python
pv.set_jupyter_backend('panel')
if 'DISPLAY' not in os.environ:
    pv.start_xvfb()
```

```python
sphere = pv.Sphere()
sphere['scalars'] = sphere.points[:, 2]
```

```python
pl = pv.Plotter(window_size=(700, 700))
pl.add_mesh(sphere)
pl.add_axes()
pl.show_bounds()
pl.show()
```
